
Esta base de datos contiene informacion sobre el indice de criminalidad en Mexico, esta base tiene como finalidad ser de apoyo para los departamentos policiales de Mexico para su buen uso, en contra de los diferentes crimenes registrados y su respectivo combate de ello.

En esta base encontraremos diversas columnas, las cuales son de gran ayuda para analisis proximos. las columnas con las cuales contamos son las siguientes :Año, Clave_Ent(Clave de Entidad), Bien juridico afectado, Tipo de delito, Subtipo de delito, Modalidad, Meses del Año, sexo, Rango de edad.
De igual forma tenemos Renglones, de los 32 estados de Mexico, de esta forma de se obtiene un facil acceso a la informacion deseada.

No obstante la base tiene valores repetidos, sin valor o mal catalogados etc... aqui mismo se lleva el proceso de limpieza el cual es explicado en cada paso.


Primeramente cargamos la base de datos

```python
import pandas as pd
df=pd.read_csv('c:/Users/Ulises/Downloads/df_sucio/df_sucio.csv')
df
```

Ahora revisamos que columnas tiene nuestra base de datos

```python
df.columns
```

Ya que sabemos que columnas tiene nuestra base, procedemos a ver si tiene regristros duplicados y de ser asi tenemos que borrarlos

```python
#Con este comando se crea un booleano que muestra renglones duplicados
df.duplicated()
```

este codigo nos muestra la suma de todos los valores duplicados

```python

df.duplicated().sum()
```

Ya conocemos nuestros valores duplicados, procedemos a borrarlos con el siguiente codigo

```python
df1=df.drop_duplicates()
df1
```

Nos aseguramos de tener todos los valores duplicados borrados

```python
df1.duplicated().sum()
```

Ya estamos seguros de que nuestra base esta limpia de valores repetidos, procedemos a filtrar columnas de nuestro interes para tener informacion mas detallada y que seran de gran ayuda

```python
df['Entidad'].value_counts()
```

Al tener el filtrado nos damos cuenta que nuestra base tiene datos inecesarios llamados "bbb" estos datos son irrelevantes y incluso peligrosos para nuestro futuro analisis  

Procedemos a borrarlos

Creamos una lista de todas nuestras columnas

```python
lista_col=df.columns
lista_col
```

Vamos a generar un ciclo for que me indique cuantas veces aparece el texto bbb en todas las columnas

```python
lista_col=df.columns
for n in lista_col:
    print(f"la columna {n} tiene de datos:")
    print(df[n].unique())
    print()
```

```python
lista_col=df.columns
for nombre in lista_col:
    print(f"En la columna {nombre} los bbb son: {df[df[nombre] == 'bbb'].shape[0]}")
```

Bien, ya filtramos nuestras columnas y podemos ver que, por lo menos en la mitad de nuestras columnas tiene el dato llamado "bbb"

Procedemos a borrar en todas nuestras columnas donde aparesca con la ayuda de un ciclo for

```python
df1=df
for i in lista_col:
    df1=df1[df1[i] != 'bbb'] #esta linea de codigo es quien borra el susodicho "bbb"
df1
```

Verificamos nuevamente nuestras columnas para asegurar que esten limpias del dato "bbb"

```python
for i in lista_col:
    print(f"En la columna {i} los bbb son: {df1[df1[i] == 'bbb'].shape[0]}")
```

En nuestra base tiene valores NaN, podriamos simplemente borrarlos pero no es deseable por que el objetivo es recuperar informacion

en su lugar los remplasaremos con valores 0 para no afectar nuestro futuro analisis 

```python
df1.fillna(value=0)
```

Ahora veremos si nuestra base necesita un cambio de valores

```python
df1.info()
```

```python
df1.head(2)
```

Como se puede observar las columnas:Febrero, Abril, Julio, Septiembre y Diciembre son de valor numerico, sin embargo en la base esta establecido como objeto, lo cual me indica de tengo que realizar un cambio de variable

En seguida hacemos uso del comando "unique" para poder visualisar el contenido que tiene la columna "Febrero"

```python
df1['Febrero'].unique()
```

Se observa que los números están entre comillas, eso indica que son caracteres y no números.

Para cambiar el tipo se ocupa el comando astype( tipo de valor)

int entero

float decimal

En este caso los convertiremos a decimal

```python

df1['Febrero']=df1['Febrero'].astype(float)
```

```python
df1['Febrero'].unique()
```

Aqui se puede apreciar como a cambiado de valor la columna Febrero de ser un objeto a un valor decimal

```python
df1.info()
```

Como bien se menciono antes que habia mas columnas con el mismo problema en valor, procedemos a realizar la misma accion para las columnas restantes

```python
df1['Abril']=df1['Abril'].astype(float)
```

```python
df1['Julio']=df1['Julio'].astype(float)
```

```python
df1['Septiembre']=df1['Septiembre'].astype(float)
```

```python
df1['Diciembre']=df1['Diciembre'].astype(float)
```

Aqui finalmente se observa que los valores acaban de cambiar a tipo numerico decimal

```python
df1.info()
```

Por ultimo ya solo queda restablecer el indice 
con ayuda del siguiente codigo

```python
df1 = df1.reset_index(drop=True)
df1
```

```python
df1=df1.fillna(value=0)
```

```python
df1.duplicated().sum()
```
