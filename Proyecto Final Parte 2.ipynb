```python

df.columns = df.columns.str.strip()


lista_meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 
               'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
meses_reales = [col for col in lista_meses if col in df.columns]

for col in meses_reales:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)


df['Target_Total_Anual'] = df[meses_reales].sum(axis=1)

print("Resultados para sustentar la Sección 7:\n" + "="*40)


print("\n[A] Variables Influyentes (Promedios más altos):")


top_delitos = df.groupby('Tipo de delito')['Target_Total_Anual'].mean().sort_values(ascending=False).head(5)
print(f"Top 5 Delitos:\n{top_delitos}")



top_entidad = df.groupby('Entidad')['Target_Total_Anual'].mean().sort_values(ascending=False).head(3)
print(f"\nTop 3 Entidades:\n{top_entidad}")




print("\n[B] Outliers y Estadísticas:")
desc = df['Target_Total_Anual'].describe(percentiles=[0.75, 0.95, 0.99])
print(f"Promedio: {desc['mean']:.2f}")
print(f"Máximo (Outlier extremo): {desc['max']}")
print(f"El 99% de los datos es menor a: {desc['99%']}")




print("\n[C] Desbalance (Sparsity):")
ceros = (df['Target_Total_Anual'] == 0).sum()
pct_ceros = (ceros / len(df)) * 100
print(f"Registros con 0 delitos: {ceros}")
print(f"Porcentaje de ceros: {pct_ceros:.2f}%")




print("\n[D] Correlación Temporal:")

if len(meses_reales) > 0:
    corr_val = df[meses_reales[0]].corr(df['Target_Total_Anual'])
    print(f"Correlación {meses_reales[0]} vs Total Anual: {corr_val:.4f}")
    


plt.figure(figsize=(10, 5))
sns.histplot(df['Target_Total_Anual'], bins=50, log_scale=(False, True)) 
plt.title('Evidencia de Outliers y Distribución (Escala Log en Y)')
plt.xlabel('Cantidad de Delitos')
plt.ylabel('Frecuencia (Log)')
plt.show()


plt.figure(figsize=(10, 5))
sns.barplot(x=top_delitos.values, y=top_delitos.index, palette='magma')
plt.title('Evidencia de Variables Influyentes (Top Delitos)')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error




X = df[['Entidad', 'Tipo de delito', 'Año']]
y = df['Target_Total_Anual']


X_encoded = pd.get_dummies(X, drop_first=True)
y_log = np.log1p(y)


print(">>> a. Dividiendo datos...")
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_log, test_size=0.2, random_state=42
)


print(">>> b. Entrenando modelo (Random Forest)...")

modelo = RandomForestRegressor(n_estimators=100, random_state=42)


modelo.fit(X_train, y_train)
print("    ¡Modelo entrenado exitosamente!")


print(">>> c. Realizando predicciones...")
y_pred_log = modelo.predict(X_test)


y_pred_real = np.expm1(y_pred_log)
y_test_real = np.expm1(y_test)


r2 = r2_score(y_test_real, y_pred_real)
mae = mean_absolute_error(y_test_real, y_pred_real)
print(f"Precisión (R2): {r2:.2f}")
print(f"Error Promedio (MAE): {mae:.0f} delitos")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error


r2 = r2_score(y_test_real, y_pred_real)
mae = mean_absolute_error(y_test_real, y_pred_real)
mse = mean_squared_error(y_test_real, y_pred_real)
rmse = np.sqrt(mse)

print("--- RESULTADOS DE LA EVALUACIÓN ---")
print(f"1. R2 Score (Precisión): {r2:.4f}")
print(f"2. MAE (Error Promedio): {mae:.2f} delitos")
print(f"3. RMSE (Error Cuadrático): {rmse:.2f} delitos")


plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test_real, y=y_pred_real, alpha=0.5, color='royalblue')


max_val = max(y_test_real.max(), y_pred_real.max())
plt.plot([0, max_val], [0, max_val], '--r', linewidth=2, label='Predicción Perfecta')

plt.title(f'Evaluación Visual: Realidad vs. Predicción (R2={r2:.2f})', fontsize=14)
plt.xlabel('Delitos Reales (Datos de Prueba)')
plt.ylabel('Delitos Predichos por el Modelo')
plt.xscale('log'); plt.yscale('log') 
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig('grafico_evaluacion_final.png')
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt




data_mapa = df.groupby('Entidad')['Target_Total_Anual'].mean().reset_index()


url_mapa = "https://raw.githubusercontent.com/angelnmara/geojson/master/mexicoHigh.json"
mapa_mexico = gpd.read_file(url_mapa)


correcciones = {
    'Mexico': 'México', 
    'Veracruz de Ignacio de la Llave': 'Veracruz',
    'Michoacán de Ocampo': 'Michoacán',
    'Coahuila de Zaragoza': 'Coahuila'
}
data_mapa['Entidad'] = data_mapa['Entidad'].replace(correcciones)


mapa_final = mapa_mexico.merge(data_mapa, left_on='name', right_on='Entidad', how='left').fillna(0)


plt.figure(figsize=(12, 8))
ax = plt.gca()

mapa_final.plot(
    column='Target_Total_Anual', 
    cmap='Reds',    
    linewidth=0.5,    
    ax=ax,            
    edgecolor='0.6',   
    legend=False    
)


plt.title('ZONA CRÍTICA: CENTRO DEL PAÍS', fontsize=24, weight='bold', color='darkred', loc='left')


plt.axis('off')

plt.tight_layout()
plt.savefig('Dash_Clean_2_Mapa.png')
plt.show()
```
