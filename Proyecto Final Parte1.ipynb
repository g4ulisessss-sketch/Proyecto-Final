```python
lista2= [
         "Bien jurídico afectado", 
         
          
         "Sexo/Averiguación previa", 
         "Rango de edad"
         ]
```

```python
for i in lista2:
    plt.figure(figsize=(10, 5))

    categoricas=df[i].value_counts().sort_index()
    
    colors = plt.cm.plasma(np.linspace(0, 1, len(categoricas)))

    


    categoricas.plot(kind='bar', color=colors)
     
    plt.title('Gráfico de Barras')
    plt.xlabel(f'{i}')
    plt.ylabel('Frecuencia')

    plt.xticks(rotation=45, ha='right')

    plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

    plt.show()
```

```python

top_10_apellido = df['Modalidad'].value_counts().nlargest(10)

colors = plt.cm.Spectral(np.linspace(0, 1, len(top_10_apellido)))

top_10_apellido.plot(kind='barh', color=colors)

plt.title('Top 10 Modalidad más Frecuentes', fontsize=16)
plt.xlabel('Frecuencia', fontsize=12)

plt.grid(True, which='both', axis='x', linestyle='--', linewidth=0.7, alpha=0.7)

plt.gca().invert_yaxis()

plt.show()
```

```python

top_10_apellido = df['Tipo de delito'].value_counts().nlargest(10)

colors = plt.cm.BrBG(np.linspace(0, 1, len(top_10_apellido)))

top_10_apellido.plot(kind='barh', color=colors)

plt.title('Top 10 Tipo de delito más Frecuentes', fontsize=16)
plt.xlabel('Frecuencia', fontsize=12)

plt.grid(True, which='both', axis='x', linestyle='--', linewidth=0.7, alpha=0.7)

plt.gca().invert_yaxis()

plt.show()
```

```python

top_10_apellido = df['Subtipo de delito'].value_counts().nlargest(10)

colors = plt.cm.RdYlBu(np.linspace(0, 1, len(top_10_apellido)))

top_10_apellido.plot(kind='barh', color=colors)

plt.title('Top 10 Subtipo de delito más Frecuentes', fontsize=16)
plt.xlabel('Frecuencia', fontsize=12)

plt.grid(True, which='both', axis='x', linestyle='--', linewidth=0.7, alpha=0.7)

plt.gca().invert_yaxis()

plt.show()
```
